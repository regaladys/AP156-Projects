## Algorithm and Results

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.special import jn
```

### Sample data: Bessel function with noise

```python
R = 256
xdata = np.linspace(0, 10, R)
data = jn(1, xdata) + 0.2*(np.random.random(R) - 0.5)

sigma = 0.3 
kernel = np.exp(-np.linspace(-5, 5, R)**2/(2*sigma**2))
kernel = np.roll(kernel, -(R/2-1))
kernel = kernel / kernel.sum()

plt.figure(1)
plt.plot(xdata, data, 'o', label="Bessel function with noise")
plt.title('Figure 1. Sample data of Bessel function with noise')
plt.show()
```

### Discrete Fourier Transform

```python
def DFT_slow(x):
    """Compute the discrete Fourier Transform of the 1D array x [3]"""
    x = np.asarray(x, dtype=float)
    N = x.shape[0]
    n = np.arange(N)
    k = n.reshape((N, 1))
    M = np.exp(-2j * np.pi * k * n / N)
    return np.dot(M, x)

def IDFT_slow(x):
    x = np.asarray(x, dtype=float)
    N = x.shape[0]
    n = np.arange(N)
    k = n.reshape((N, 1))
    M = np.exp(2j * np.pi * k * n / N)
    return np.dot(M, x)/N
```

```python
import cmath
# Discrete fourier transform
def dft(x):
    t = []
    N = len(x)
    for k in range(N):
        a = 0
        for n in range(N):
            a += x[n]*cmath.exp(-2j*cmath.pi*k*n*(1/N))
        t.append(a)
    return t
# Inverse discrete fourier transform
def idft(t):
    x = []
    N = len(t)
    for n in range(N):
        a = 0
        for k in range(N):
            a += t[k]*cmath.exp(2j*cmath.pi*k*n*(1/N))
        a /= N
        x.append(a)
    return x
```

### DFT and FFT

```python
dftdata = DFT_slow(data)
dftkernel = DFT_slow(kernel)
smoothd = IDFT_slow(dftdata*dftkernel)
smoothf = np.fft.ifft(np.fft.fft(data)*np.fft.fft(kernel))

plt.figure(1)
plt.plot(xdata, data, 'o', label="Bessel function with noise")
plt.plot(xdata, smoothd, lw=3, label="smooth data DFT")
#plt.plot(xdata, idft(dft(data)), lw=3, label="smooth data by DFT method 2")
plt.plot(xdata, smoothf, lw=3, label="smooth data by FFT")
#legend = plt.legend(loc='upper center', shadow=False)
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.title('Figure 2. Sample data and smooth data by DFT')
plt.show()
```
